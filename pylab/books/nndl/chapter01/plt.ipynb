```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid_d(x):
    y = 1 / (1 + np.exp(-x))
    return y * (1 - y)


def plot_sigmoid():
    x = np.arange(-4, 4, 0.1)
    y = sigmoid_d(x)
    plt.plot(x, y)
#     plt.title("Sigmoid 导数函数图像")
    plt.show()
```

```python
plot_sigmoid()
```
