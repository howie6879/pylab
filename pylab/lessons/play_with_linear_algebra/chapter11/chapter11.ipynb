# 坐标转换和线性变换

## 空间的基和坐标系

在n维空间，如果给定一组基，任何一个向量(或者是点) 都可以表示成这组基的线性组合且表示方法唯一。

<img style="margin-left:0%" src="https://gitee.com/howie6879/oss/raw/master/uPic/Jc1oPr.png" width="50%">

$$
\begin{array}{l}
\left(\begin{array}{ll}
4 & 2 \\
1 & 3
\end{array}\right)\left(\begin{array}{l}
2 \\
2
\end{array}\right)=\left(\begin{array}{c}
12 \\
8
\end{array}\right) \\
=2\left(\begin{array}{l}
4 \\
1
\end{array}\right)+2\left(\begin{array}{l}
2 \\
3
\end{array}\right)
\end{array}
$$

## 任意坐标系转换

问题：假设有一组基：$B=\left\{\vec{b}_{1}, \vec{b}_{2}, \vec{b}_{3}, \ldots, \vec{b}_{n}\right\}$，另一组基：$C=\left\{\vec{c}_{1}, \vec{c}_{2}, \vec{c}_{3}, \ldots, \bar{c}_{n}\right\}$，已知$[\vec{x}]_{B}$，求：$[\vec{x}]_{C}$？

以标准坐标系为桥梁即可解决：
$$
{[\vec{x}]_{\varepsilon} = P_{B}[\vec{x}]_{B}} \\
{[\vec{x}]_{C} = P_{C}^{-1}[\vec{x}]_{\varepsilon}}\\
{[\vec{x}]_{C} = P_{C}^{-1}P_{B}[\vec{x}]_{B}}
$$

扩展：

$$
\begin{array}{l}
{[\vec{x}]_{C}=P_{C}^{-1} P_{B}[\bar{x}]_{B} \quad[\vec{x}]_{C}=P_{B->C}[\bar{x}]_{B} \quad P_{B->C}=P_{C}^{-1} P_{B}} \\
P_{C}[\vec{x}]_{C}=P_{B}[\vec{x}]_{B} \\
P_{B}^{-1} P_{C}[\vec{x}]_{C}=[\vec{x}]_{B} \quad[\vec{x}]_{B}=P_{C \rightarrow B}[\bar{x}]_{C} \quad P_{C \rightarrow B}=P_{B}^{-1} P_{C}
\end{array}
$$

## 线性变换

一个变换T(x)称为线性变化，必须满足:

$$
\begin{array}{l}
T(u+v)=T(u)+T(v) \\
T(c u)=c T(u) \quad c \in R
\end{array}
$$

一个矩阵，既可以看作是空间，又可以看做是变换。我们可以把矩阵看作是向量的函数$A\vec{x}$，矩阵所表示的变换，均为线性变换。
