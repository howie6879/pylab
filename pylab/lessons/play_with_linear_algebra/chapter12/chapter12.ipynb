# 行列式

矩阵可以表示一组向量，方阵表示n个n维向量，行列式是方阵的一个属性，行列式表示向量组在空间中形成的有向体积。 

关于计算：

$$
\begin{array}{l}
\operatorname{det}\left(\begin{array}{ll}
a & b \\
c & d
\end{array}\right)=a d-b c \\
\left|\begin{array}{ll}
a & b \\
c & d
\end{array}\right|=a d-b c
\end{array}
$$

## 行列式的四大基本性质

- 交换行列式的两行，则行列式的值取反
- $\operatorname{det} I=1$
- 方阵的**某一行**乘以一个数k，则其对应的行列式也缩放了k倍：
$$
\left|\begin{array}{cc}
k a & k b \\
c & d
\end{array}\right|=k\left|\begin{array}{ll}
a & b \\
c & d
\end{array}\right|
$$
- 方阵的某一行加上一行数，则有:
$$
\left|\begin{array}{cc}
a+a^{\prime} & b+b^{\prime} \\
c & d
\end{array}\right|=\left|\begin{array}{ll}
a & b \\
c & d
\end{array}\right|+\left|\begin{array}{cc}
a^{\prime} & b^{\prime} \\
c & d
\end{array}\right|
$$

## 行列式与矩阵的逆

- 如果行列式有一行为0，则行列式的值为0
- 如果行列式的两行相同，则行列式的值为0
- 如果行列式的一行是另一行的k倍，则行列式的值为0
- !!**如果行列式的一行是其他行的线性组合，则行列式的值为0**

最终得出，如果这组向量线性相关，则行列式的值为0 <----> 矩阵不可逆

## 计算行列式的值

- 如果一个行列式的一行加(减)另一行的k倍，行列式的值不变：

$$
\begin{aligned}
\left|\begin{array}{lll}
a & b & c \\
d & e & f \\
g & h & i
\end{array}\right| &=\left|\begin{array}{ccc}
a & b & c \\
d & e & f \\
g-k a & h-k b & i-k c
\end{array}\right| \\
&=\left|\begin{array}{lll}
a & b & c \\
d & e & f \\
g & h & i
\end{array}\right|-\left|\begin{array}{ccc}
a & b & c \\
d & e & f \\
k a & k b & k c
\end{array}\right|=\left|\begin{array}{lll}
a & b & c \\
d & e & f \\
g & h & i
\end{array}\right|
\end{aligned}
$$
- 对角矩阵的行列式
$$
\begin{array}{c}
\left|\begin{array}{cccc}
d_{1} & 0 & \ldots & 0 \\
0 & d_{2} & \ldots & 0 \\
\ldots & \ldots & \ldots & \ldots \\
0 & 0 & \ldots & d_{n}
\end{array}\right| \\
=d_{1} d_{2} \ldots d_{n}
\end{array}
$$
- 上三角矩阵的行列式
$$
\left|\begin{array}{cccc}
d_{1} & a_{12} & \ldots & a_{1 n} \\
0 & d_{2} & \ldots & a_{2 n} \\
\ldots & \ldots & \ldots & \ldots \\
0 & 0 & \ldots & d_{n}
\end{array}\right| \\
=d_{1} d_{2} \ldots d_{n}
$$
- 下三角矩阵的行列式
$$
\left|\begin{array}{cccc}
d_{1} & 0 & \ldots & 0 \\
a_{21} & d_{2} & \ldots & 0 \\
\ldots & \ldots & \ldots & \ldots \\
a_{n 1} & a_{n 2} & \ldots & d_{n}
\end{array}\right| \\
=d_{1} d_{2} \ldots d_{n}
$$

## 初等矩阵与行列式

$\operatorname{det}(A \cdot B)=\operatorname{det}(A) \cdot \operatorname{det}(B)$：
- A或B中的某一行和其他行线性相关
- A和B中的所有行都线性无关：引入初等矩阵来证明
$$
\begin{array}{l}
\operatorname{det}(A \cdot B)=\operatorname{det}(A) \cdot \operatorname{det}(B) \\
\operatorname{det}\left(A \cdot A^{-1}\right)=\operatorname{det}(A) \cdot \operatorname{det}\left(A^{-1}\right) \\
\operatorname{det}(I)=\operatorname{det}(A) \cdot \operatorname{det}\left(A^{-1}\right) \\
\operatorname{det}(A) \cdot \operatorname{det}\left(A^{-1}\right)=1 \\
\operatorname{det}\left(A^{-1}\right)=\frac{1}{\operatorname{det}(A)}
\end{array}
$$ 
- 如果E表示两行交换位置：$\operatorname{det}(E)=\operatorname{det}\left(E^{T}\right)$

## 行式就是列式

形式化表达：$\operatorname{det}(A)=\operatorname{det}\left(A^{T}\right)$

证明思路：任意$A$可以分解成$P L U P^{\prime}$：$\operatorname{det}(A)=\operatorname{det}\left(P L U P^{\prime}\right)$

$$
\begin{array}{l}
\operatorname{det}(A)=\operatorname{det}\left(P L U P^{\prime}\right)=\operatorname{det}(P) \cdot \operatorname{det}(L) \cdot \operatorname{det}(U) \cdot \operatorname{det}\left(P^{\prime}\right) \\
\operatorname{det}\left(A^{T}\right)=\operatorname{det}\left(\left(P L U P^{\prime}\right)^{T}\right)=\operatorname{det}\left(P^{\prime T} U^{T} L^{T} P^{T}\right) \\
=\operatorname{det}\left(P^{\prime T}\right) \operatorname{det}\left(U^{T}\right) \operatorname{det}\left(L^{T}\right) \operatorname{det}(P^{T})
\end{array}
$$
